```python
import tweepy #https://github.com/tweepy/tweepy
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import time
import json
```

```python
df2 = pd.read_csv("../data/billboard.csv", sep='|', names=['date','pos','song','long_artist'])
df2["artist"] = df2["long_artist"].map(lambda x : x.split('featuring')[0])
df2["ID"] = df2["song"] + "%" +  df2["artist"]
```

```python
df= df2.pivot('ID', 'date', 'pos')
df["ID"] = df.index
df["song"] = df["ID"].map(lambda x : x.split('%')[0])
df["artist"] = df["ID"].map(lambda x : x.split('%')[1])
df.fillna(101,inplace=True)
df.index = range(1,len(df)+1)
df["IDN"] = df.index
df.shape
```

```python
Input = [ ["#"+str(id.split('%')[0]), str(id.split('%')[1])]  for id in df["ID"]]
```

```python
#Twitter API credentials
consumer_key = os.getenv("TWITTER_CONSUMER_KEY")
consumer_secret = os.getenv("TWITTER_CONSUMER_SECRET")
access_key = os.getenv("TWITTER_ACCESS_TOKEN")
access_secret = os.getenv("TWITTER_ACCESS_TOKEN_SECRET")
```

```python
#authorize twitter, initialize tweepy
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api = tweepy.API(auth)
```

```python
from pymongo import MongoClient
client = MongoClient('localhost', 27017)
```

## Cursor

```python
def twitter_scrape(query, tab):
    for tweet in tweepy.Cursor(api.search, q=query).items():
        tweet_json= tweet._json
        tab.insert(tweet_json)  
```

```python
def continuous_scraper(query, tab):
    '''
    Helper function to use with a list of screen names. It will run exhaustively through list
    skipping problematic accounts (banned/suspended/etc)
    '''
    try:
        twitter_scrape(query, tab)
    except KeyboardInterrupt:
        print "KeyboardInterrupt"
        return
    except:
        print "Error encountered at {} on {}, will try in 15 minutes. The curent number of data is {} ".format\
        (time.asctime(time.localtime()), query[0], tab.count())
        time.sleep(901)
        print "twitter scrape resumes"
        continuous_scraper(query, tab)
        return
```

```python
# Access/Initiate Database
db = client['song3_database']

MaxIt = df.shape[0]

for i in range(144,MaxIt):
    print "Start scraping for id = {}".format(i)
    table_name = "test_01_" + str(i)
    tab = db[table_name]
    query = Input[i]
    continuous_scraper(query, tab)
    print "Final number of the data is {} at id = {} among {}".format(tab.count(),i, MaxIt)
    print "---------------------------------------------------"
    
print "Work is completed"
```
