```python
import tweepy
import os
import time
from datetime import datetime, timedelta
import tweepy
from pymongo import MongoClient
import json
#from tweepy.streaming import StreamListener
#from tweepy import Stream



```

```python
#Twitter API credentials
consumer_key = os.getenv("TWITTER_CONSUMER_KEY")
consumer_secret = os.getenv("TWITTER_CONSUMER_SECRET")
access_key = os.getenv("TWITTER_ACCESS_TOKEN")
access_secret = os.getenv("TWITTER_ACCESS_TOKEN_SECRET")
```

```python
#authorize twitter, initialize tweepy
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api = tweepy.API(auth)
```

```python
client = MongoClient('localhost', 27017)
api = tweepy.API(auth)
db = client['rest_test2_database']
tweets_db = db['rest_test2_table']
```

```python
query = ["JUSTIN BIEBER", "#work"]
#query = ["JUSTIN BIEBER"]
```

```python
search_results = tweepy.Cursor(api.search, q=query).items()
```

```python
for tweet in search_results:
    tweet_json = json.loads(tweet)
    tweets_db.insert(tweet_json)
```

```python
while 1:

    #max_tweets = 20000

    search_results = tweepy.Cursor(api.search, q=query).items()
    for tweet in search_results:
        tweet_json = json.loads(tweet)
        tweets_db.insert(tweet_json)
        
    dt = datetime.now() + timedelta(minutes=16)
    # dt = dt.replace(minute=10)

    while 1:
        if datetime.now() > dt:
            break

        time.sleep(950)
```
